# Pearson Correlation

```python
ys = data.columns
ys = ys.drop('convergence_speed')
ys = ys.drop('convergence_iterations')
ys = ys.drop('matrix')
ys = ys.drop('sle')
xs = ['convergence_speed', 'convergence_iterations']
COLUMNS = ['number_of_nodes', 'number_of_edges', 'diameter', 'radius',
           'average_deg', 'max_in_deg', 'max_out_deg', 'norm_sle']
d = []
for x in xs:
    row = []
    for y in ys:
        pearson = pearsonr(data[x], data[y])
        row.append(pearson.statistic)
    d.append(row)
pearson_df = pd.DataFrame(d, index=xs, columns=ys)
```
