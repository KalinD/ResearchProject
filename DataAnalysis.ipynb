# Analysing the results

```python
# %pip install plotly
# %pip install matplotlib

import math
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('./matrices_final.csv')
data2 = pd.read_csv('./matrices2_final.csv')
data = pd.concat([data1, data2])
data.drop('Unnamed: 0', axis=1, inplace=True)
data.head(10)
```

```python
data.head()
```

## Max in-degree

```python
plt.scatter(x=data['max_in_deg'], y=data['convergence_speed'])
plt.show()
```

```python
plt.scatter(x=data['max_in_deg'], y=data['convergence_iterations'])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x=data['max_in_deg'], y=data['convergence_speed'])

# Set logarithmic scale on the x variable
ax.set_xscale("log")
ax.set_yscale("log")
```

```python
fig, ax = plt.subplots()
ax.scatter(x=data['max_in_deg'], y=data['convergence_iterations'])

# Set logarithmic scale on the x variable
ax.set_xscale("log")
ax.set_yscale("log")
```
